```python
import sqlite3
import pandas as pd
db = 'sales.db'

#Run queries on database
def run_query(my_query):
    with sqlite3.connect(db) as conn:
        return pd.read_sql(my_query, conn)

#Execute queries on database
def run_command(my_command):
    with sqlite3.connect(db) as conn:
        conn.isolation_level = None
        conn.execute(my_command)
    print('✔️ SUCCESS')
    
#Insert a dataframe into a table
def insert_dataframe(df, table_name):
    
    with sqlite3.connect(db) as conn:
        table_columns = run_query(f'select * from {table_name}').columns
        df.columns = table_columns
        df.to_sql(table_name, conn, if_exists='append', index=False)
   
    print('✔️ SUCCESS')
```

```python
create_employees_table_query = '''
CREATE TABLE IF NOT EXISTS employees (
    Id INTEGER PRIMARY KEY,
    employee_number INTEGER,
    last_name TEXT,
    first_name TEXT,
    line_manager TEXT,
    title TEXT
);
'''

run_command(create_employees_table_query)

print("✔️ Employees table created successfully.")
```

```python
create_customers_table_query = '''
CREATE TABLE IF NOT EXISTS customers (
    id INTEGER PRIMARY KEY,
    first_name TEXT,
    last_name TEXT,
    country TEXT,
    phone_number TEXT,
    sales_contact_code TEXT
);
'''

run_command(create_customers_table_query)

print("✔️ Customers table created successfully.")
```

```python
create_invoice_items_table_query = '''
CREATE TABLE IF NOT EXISTS invoice_items (
    id INTEGER PRIMARY KEY,
    invoice_id INTEGER,
    product_code TEXT,
    quantity INTEGER
);
'''
run_command(create_invoice_items_table_query)

print("✔️ Invoice Items table created successfully.")
```

```python
create_invoices_table_query = '''
CREATE TABLE IF NOT EXISTS invoices (
    id INTEGER PRIMARY KEY,
    invoice_number TEXT,
    invoice_date TEXT,
    billing_phone TEXT,
    total REAL
);
'''
run_command(create_invoices_table_query)

print("✔️ Invoices table created successfully.")
```

```python
import sqlite3
import pandas as pd

db = 'sales.db'

conn = sqlite3.connect(db)

query = '''
SELECT
    sales_contact_code AS SalesOfficerCode,
    COUNT(*) AS TotalCustomers
FROM
    customers
GROUP BY
    sales_contact_code
ORDER BY
    TotalCustomers DESC;
'''

performance_df = pd.read_sql(query, conn)

conn.close()

print(performance_df)
```

```python
import sqlite3

# Define the database file
db_file = 'sales.db'

# Establish a connection to the SQLite database
conn = sqlite3.connect(db_file)
cursor = conn.cursor()

# Define the SQL query to drop the view if it already exists
drop_view_query = 'DROP VIEW IF EXISTS product_sales_summary;'

# Execute the query to drop the view
cursor.execute(drop_view_query)

# Define the SQL query to create the view
create_view_query = '''
CREATE VIEW product_sales_summary AS
SELECT
    ii.product_code,
    SUM(ii.quantity) AS total_quantity,
    SUM(ii.quantity * p.unit_price) AS total_amount
FROM
    invoice_items AS ii
JOIN
    products AS p ON ii.product_code = p.product_code
GROUP BY
    ii.product_code;
'''

# Execute the query to create the view
cursor.execute(create_view_query)

# Commit the transaction
conn.commit()

# Close the connection
conn.close()
```

```python
import sqlite3

# Define the database file
db_file = 'sales.db'

# Close any existing connection
try:
    conn.close()
except NameError:
    pass  # No existing connection to close

# Establish a new connection to the SQLite database
conn = sqlite3.connect(db_file)
cursor = conn.cursor()

# Drop the existing invoice_items table if it exists
cursor.execute("DROP TABLE IF EXISTS invoice_items")

# Create the invoice_items table with the correct schema
cursor.execute('''
CREATE TABLE IF NOT EXISTS invoice_items (
    id INTEGER PRIMARY KEY,
    product_code TEXT,
    quantity INTEGER
)
''')

# Commit the transaction
conn.commit()

# Insert sample data into the invoice_items table
invoice_items_data = [
    (1, 'P001', 5),
    (2, 'P002', 10),
    (3, 'P001', 3),
    # Add more sample data as needed
]
cursor.executemany('INSERT INTO invoice_items VALUES (?, ?, ?)', invoice_items_data)

# Drop the existing products table if it exists
cursor.execute("DROP TABLE IF EXISTS products")

# Create the products table with the correct schema
cursor.execute('''
CREATE TABLE IF NOT EXISTS products (
    id INTEGER PRIMARY KEY,
    product_code TEXT,
    product_name TEXT,
    unit_price REAL
)
''')

# Insert sample data into the products table
products_data = [
    (1, 'P001', 'Product 1', 10.99),
    (2, 'P002', 'Product 2', 20.99),
    # Add more sample data as needed
]
cursor.executemany('INSERT INTO products VALUES (?, ?, ?, ?)', products_data)

# Commit the transaction
conn.commit()

# Close the connection
conn.close()
```
