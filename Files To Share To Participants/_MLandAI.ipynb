```python
def outliers_bounds(df, column_name):

    values = df[column_name]
    q1 = np.percentile(values, 25)
    q3 = np.percentile(values, 75)
    iqr = q3 -q1
    co = iqr * 1.5
    lower_bound = q1 - co
    upper_bound = q3 + co

    return lower_bound, upper_bound
```

```python
def create_dummies(df, numerical_columns, dummy_columns):

    numerical_df = df[numerical_columns]
    dummy_df = pd.get_dummies(df[dummy_columns], dtype = int)
    
    numerical_dummies = pd.concat([numerical_df, dummy_df], axis = 1)
    
    return numerical_dummies
```

# Custom function to train a model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
def train_and_evaluate_model(model, train_features, train_target, test_features, test_target, task='classification', evaluation_metrics=['accuracy']):
    model.fit(train_features, train_target)
    # Make predictions
    predictions = model.predict(test_features)
    evaluation_results = {}
    # Evaluate using specified metrics
    if task == 'classification':
        for metric in evaluation_metrics:
            if metric == 'accuracy':
                evaluation_results['accuracy'] = accuracy_score(test_target, predictions)
            elif metric == 'precision':
                evaluation_results['precision'] = precision_score(test_target, predictions, average='weighted')
            elif metric == 'recall':
                evaluation_results['recall'] = recall_score(test_target, predictions, average='weighted')
            elif metric == 'f1':
                evaluation_results['f1'] = f1_score(test_target, predictions, average='weighted')
            # Add more classification metrics as needed
    elif task == 'regression':
        for metric in evaluation_metrics:
            if metric == 'mse':
                evaluation_results['mse'] = mean_squared_error(test_target, predictions)
            elif metric == 'mae':
                evaluation_results['mae'] = mean_absolute_error(test_target, predictions)
            elif metric == 'r2':
                evaluation_results['r2'] = r2_score(test_target, predictions)
    return model, predictions, evaluation_results
```
